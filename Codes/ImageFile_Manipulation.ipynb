```python
fileNames=[]
fileFolders=[]
imageDir=[]
imageLabels=[]
for subdir, dirs,i in os.walk(trainingDigits):
    fileNames.append(i)
    fileFolders.append(subdir)
    for k in i:
        file=subdir+'/'+k
        if (len(k.split('.')[0])==1):
            print (file)
            imageLabels.append('0'+k.split('.')[0])
            imageDir.append(file)
        else:
            imageLabels.append(k.split('.')[0])
            imageDir.append(file)
        #print (k.split('.')[0])
```
