```python
import cv2
import os
import random
#import matplotlib.pylab as plt
import matplotlib.pyplot as plt
from glob import glob
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
```

```python
trainingDigits="/home/berhe/Desktop/Geni_Project//Hand-written-digit-Recognition-Project/Data collected"
```

```python
fileNames=[]
fileFolders=[]
imageDir=[]
imageLabels=[]
for subdir, dirs,i in os.walk(trainingDigits):
    fileNames.append(i)
    fileFolders.append(subdir)
    for k in i:
        file=subdir+'/'+k
        imageLabels.append(k.split('.')[0])
        imageDir.append(file)
```
